5.What are iterators in Python and how do they differ from iterables?

->In Python, an iterator is an object that implements the iterator protocol, which consists of two methods: iter() and next(). An iterator allows you to traverse through a collection of elements (like lists, tuples, or dictionaries) one element at a time, without needing to know the underlying structure of the collection.

The iter() method returns the iterator object itself and is required to make the object iterable.
The next() method returns the next value from the iterator. When there are no more items to return, it raises a StopIteration exception to signal that the iteration is complete.

How do Iterators Differ from Iterables?

Iterable: An iterable is any Python object that can return its elements one at a time. Examples of iterables include lists, tuples, strings, and dictionaries. An iterable must implement the iter() method, which returns an iterator.

Iterator: An iterator is a specific type of iterable that keeps track of its current position during iteration. It implements both the iter() and next() methods. While you can iterate over an iterable multiple times, an iterator can only be traversed once. After it has been exhausted, you cannot reuse it without creating a new iterator.

Example

Here’s an example to illustrate the difference between an iterable and an iterator:

```python
# Example of an iterable (a list)
my_list = [1, 2, 3]

# Getting an iterator from the iterable
my_iterator = iter(my_list)

# Using the iterator to access elements
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3

# The next call will raise StopIteration
# print(next(my_iterator))  # Uncommenting this line will raise StopIteration

# You can create a new iterator from the iterable
new_iterator = iter(my_list)
print(next(new_iterator))  # Output: 1 (new iterator starts from the beginning)

''' In this example: - 'my_list' is an iterable because it can be iterated over.'my_iterator' is an iterator created from 'my_list'. It allows us to access the elements of the list one by one using the next() function.'''
```

6.Explain the concept of generators in Python and how they are defined.

->Generators in Python are special functions that return an iterator, generating values on-the-fly using the yield keyword instead of returning values with return. This allows them to be memory efficient and suitable for handling large datasets. They can be defined similarly to regular functions but utilize yield to return values sequentially while maintaining state for subsequent calls. Additionally, generators can be created using generator expressions, which use a concise syntax akin to list comprehensions but with parentheses. The key benefits include memory efficiency and the ability to work with streams of data lazily.

Example of a Generator:
Here’s a simple example of a generator that produces a sequence of numbers:
python


def count_up_to(n):
    
    count = 1
    while count <= n:
        yield count  # Yield the current count
        count += 1   # Increment the count

# Using the generator
counter = count_up_to(5)

# Iterating through the generator


for number in counter:   
    print(number)

Explanation of the Example
Generator Function: The function count_up_to(n) is defined as a generator. It uses a while loop to yield numbers from 1 to n.

Yield Statement: Each time the yield statement is executed, the current value of count is returned, and the function's state is saved.

Iteration: When we create a generator object by calling count_up_to(5), it does not execute the function immediately. Instead, we can iterate over it using a for loop, which calls next() on the generator until it is exhausted.


7.What are the advantages of using generators over regular functions?

->The advantages of using generators over regular functions include improved memory efficiency since generators yield values one at a time, allowing them to handle large datasets without using excessive memory. They maintain their execution state, allowing resumption after yielding, which regular functions do not support. Generators can also simplify code through generator expressions, which reduce boilerplate. Lastly, they enable performance enhancements by generating items on demand rather than storing entire collections in memory.



8.What is a lambda function in Python and when is it typically used?

->A lambda function in Python is a small, anonymous function defined using the lambda keyword. It can take any number of arguments but can only have one expression. Lambda functions are typically used for short, simple operations, especially as arguments for higher-order functions like map(), filter(), and sorted().

Example:
Here’s a simple example of a lambda function that adds two numbers:

add = lambda x, y: x + y

result = add(5, 3)

print(result) # Output: 8

In this example, add is a lambda function that takes two arguments, x and y, and returns their sum. You can call this function just like a regular function.

Usage in Higher-Order Functions:
Lambda functions are often used with functions like map() or filter(). For instance, if you want to square each number in a list, you can use a lambda function with map():

numbers = [1, 2, 3, 4]

squared_numbers = list(map(lambda x: x ** 2, numbers))

print(squared_numbers)  # Output: [1, 4, 9, 16]

In this case, the lambda function lambda x: x ** 2 is applied to each element in the numbers list, resulting in a new list of squared values.

10.What is the difference between map(), reduce(), and filter() functions in Python?

->Certainly! Let's clarify the differences between map(), reduce(), and filter() in Python with supporting examples for each.

1. map():
The map() function applies a specified function to each item in an iterable (like a list) and returns a new iterable with the results.Example:

# Function to square a number
def square(x):
    return x * x

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Using map to apply the square function
squared_numbers = list(map(square, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

2. filter():The filter() function applies a specified function to each item in an iterable and returns a new iterable containing only the items for which the function returns True.

Example:
# Function to check if a number is even
def is_even(x):
    
    return x % 2 == 0

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Using filter to get only even numbers
even_numbers = list(filter(is_even, numbers))

print(even_numbers)  # Output: [2, 4]

3. reduce():The reduce() function (from the functools module) applies a function of two arguments cumulatively to the items of an iterable, reducing it to a single output.

#Example:
from functools import reduce

#Function to add two numbers
def add(x, y):

    return x + y

#List of numbers
numbers = [1, 2, 3, 4, 5]

#Using reduce to sum the numbers
total = reduce(add, numbers)

print(total)   # Output: 15

Summary:

* map() transforms each item in an iterable and returns a new iterable with the transformed items.
* filter() selects items from an iterable based on a condition and returns a new iterable with the selected items.

* reduce()combines items in an iterable into a single cumulative result using a specified function.

```python
#1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_evens(numbers):
    # Initialize the sum to 0
    sum = 0

    # Loop through the list of numbers
    for num in numbers:
        # Check if the number is even
        if num % 2 == 0:
            sum += num  # Add the even number to the sum

    return sum


# Example usage
numbers = [1, 2, 3, 4, 7, 6,10]
print(sum_of_evens(numbers))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage
print(is_prime(11))
print(is_prime(4))
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n_terms:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a
        else:
            raise StopIteration

# Example usage
fib = Fibonacci(10)
for num in fib:
    print(num, end=" ")
```

```python
# 6.Write a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage
for power in powers_of_2(5):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_list = [0, 20, 37, 100]
fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))
print(fahrenheit_list)
```

```python
# 10.Create a Python program that uses `filter()` to remove all the vowels from a given string
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))

# Example usage
string = "hello world"
print(remove_vowels(string))
```

```python
"""11.Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.
Write a Python program using lambda and map."""
orders = [
    [101, 4,40.00],
    [202, 5, 56.80],
    [302, 3, 32.95],
    [402, 3, 24.99],
]
def calculate_order(order):
    order_number, price_per_item, quantity = order
    total_price = price_per_item * quantity
    if total_price < 100:
        total_price += 10
    return (order_number, total_price)
ans = list(map(lambda order: calculate_order(order), orders))

print(ans)
```
